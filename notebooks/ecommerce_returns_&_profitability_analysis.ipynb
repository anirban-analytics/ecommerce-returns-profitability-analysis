### ⏱️ Feature Engineering: Delivery and Approval Metrics

In this step, we create new features to capture key time-based performance metrics in the order lifecycle.

- **delivery_days**: Calculates the total number of days taken to deliver an order from the purchase date.
- **approval_delay**: Measures the time taken for order approval after purchase.
- **late_delivery**: A binary indicator that flags whether an order was delivered later than the estimated delivery date.

These engineered features are essential for analyzing operational efficiency, delivery performance, and customer satisfaction.

```python
df["delivery_days"] = (
    df["order_delivered_customer_date"] -
    df["order_purchase_timestamp"]
).dt.days

df["approval_delay"] = (
    df["order_approved_at"] -
    df["order_purchase_timestamp"]
).dt.days

df["late_delivery"] = (
    df["order_delivered_customer_date"] >
    df["order_estimated_delivery_date"]
).astype(int)
```

### ✅ Output Interpretation

New columns have been successfully added to the dataset:

- **delivery_days** and **approval_delay** enable quantitative analysis of time delays in the order process.
- **late_delivery** provides a clear indicator for identifying delayed deliveries, which can be used for performance tracking and predictive modeling.

These features enhance the dataset by enabling deeper insights into logistics efficiency and service quality.

### 💰 Feature Engineering: Financial Metrics

In this step, we create additional features to capture key financial aspects of each transaction.

- **total_order_value**: Represents the total cost of an order, including product price and freight charges.
- **freight_ratio**: Measures the proportion of freight cost relative to the product price, helping assess shipping cost impact.
- **value_per_installment**: Calculates the average payment amount per installment for orders paid in multiple parts.

These features are important for analyzing revenue structure, cost distribution, and customer payment behavior.

```python
df["total_order_value"] = df["price"] + df["freight_value"]

df["freight_ratio"] = df["freight_value"] / df["price"]

df["value_per_installment"] = (
    df["payment_value"] / df["payment_installments"]
)
```

### ✅ Output Interpretation

The new financial features have been successfully added to the dataset:

- **total_order_value** provides a complete view of transaction value.
- **freight_ratio** helps identify orders with disproportionately high shipping costs.
- **value_per_installment** offers insight into customer payment patterns, especially for installment-based purchases.

These enhancements enable more detailed financial analysis and support advanced business insights related to pricing, logistics, and payment trends.

### ⚠️ Feature Engineering: Return Risk Indicator

In this step, we create a new feature called **return_risk** to identify potentially problematic orders.

This feature is based on two key conditions:
- Orders with **low review scores (≤ 2)**, indicating customer dissatisfaction
- Orders with **late delivery**, reflecting poor service experience

By combining these conditions, we create a binary indicator to flag orders that are more likely to result in returns or negative outcomes.

```python
df["return_risk"] = (
    (df["review_score"] <= 2) |
    (df["late_delivery"] == 1)
).astype(int)
```

### ✅ Output Interpretation

The **return_risk** column has been successfully created as a binary variable:

- **1** → High risk (low review score or late delivery)
- **0** → Low risk

This feature is particularly valuable for predictive modeling and risk analysis, as it helps identify orders that may lead to returns, complaints, or customer dissatisfaction.

### ⚠️ Return Risk Analysis by Customer State

In this step, we calculate the return risk percentage across different customer states.

The `return_risk` flag (derived earlier) identifies potentially problematic orders based on:
- Low review scores (≤ 2)
- Late deliveries

We use conditional aggregation to compute the percentage of high-risk orders in each state.

```python
query = """
SELECT
    customer_state,
    COUNT(*) FILTER (WHERE return_risk = 1) * 100.0 / COUNT(*) AS return_rate_pct
FROM olist_master
GROUP BY customer_state
ORDER BY return_rate_pct DESC
LIMIT 10;
"""

pd.read_sql(query, engine)
```

### 💸 Revenue Loss Analysis by Product Category

In this step, we analyze revenue loss due to high return-risk orders across product categories.

We calculate:
- Returned revenue (from high-risk orders)
- Total revenue
- Percentage revenue loss (`loss_pct`)

This helps identify which categories are most impacted by returns and operational inefficiencies.

```python
query = """
SELECT
    product_category_name_english,
    ROUND(SUM(price) FILTER (WHERE return_risk=1)::numeric, 2) AS returned_revenue,
    ROUND(SUM(price)::numeric, 2) AS total_revenue,
    ROUND(
        (
            SUM(price) FILTER (WHERE return_risk=1)::numeric
            / SUM(price)::numeric
        ) * 100, 2
    ) AS loss_pct
FROM olist_master
GROUP BY product_category_name_english
ORDER BY returned_revenue DESC
LIMIT 10;
"""

pd.read_sql(query, engine)
```

- **Computers & Accessories (~22.9%)** and **Bed, Bath & Table (~22.6%)** also exhibit significant loss ratios.
- Some categories like **CDs, DVDs & Musicals** show `NaN` values, likely due to:
  - No return-risk orders

#### 📊 Trend Analysis:
- High-revenue categories are also experiencing **high return-related losses**
- Certain categories have **disproportionately high loss percentages**, indicating deeper operational or product issues

#### 💡 Business Insights:

### 🚚 Impact of Delivery Speed on Return Risk

In this step, we categorize orders based on delivery speed using `delivery_days`:

- Fast (≤ 3 days)
- Moderate (4–7 days)
- Slow (8–14 days)
- Very Slow (> 14 days)

We then analyze how return risk varies across these delivery speed categories to understand the impact of logistics performance on customer satisfaction and returns.

```python
query = """
SELECT
    CASE
        WHEN delivery_days <= 3 THEN 'Fast'
        WHEN delivery_days <= 7 THEN 'Moderate'
        WHEN delivery_days <= 14 THEN 'Slow'
        ELSE 'Very Slow'
    END AS delivery_speed,

    COUNT(*) AS orders,
    ROUND(
        COUNT(*) FILTER (WHERE return_risk=1) * 100.0
        / COUNT(*), 2
    ) AS return_rate_pct

FROM olist_master
WHERE delivery_days IS NOT NULL
GROUP BY delivery_speed
ORDER BY return_rate_pct DESC;
"""

pd.read_sql(query, engine)
```

### 📈 Key Insights & Business Interpretation

#### 🔍 Observations:
- **Very Slow deliveries (>14 days)** have the highest return rate (~36.86%), which is significantly higher than all other categories.
- Return rate drops sharply as delivery speed improves:
  - Slow: ~12.54%
  - Moderate: ~10.72%
  - Fast: ~9.36%
- The number of orders is also highest in the **Slow and Very Slow categories**, amplifying the overall business impact.

#### 📊 Trend Analysis:
- There is a **strong positive correlation between delivery time and return risk**
- As delivery delays increase, customer dissatisfaction and return probability increase drastically
